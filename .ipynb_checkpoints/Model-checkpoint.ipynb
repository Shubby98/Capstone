```python
# Read Captions
def readcaptions(path):
    f = open(path)
    text = f.read()
    f.close()
    captions = {}
    for line in text.split("\n")[:-1]:
        pid , cap = line.split("\t")
        pid = pid[:-2]
        if(captions.get(pid)==None):
            captions[pid] = [cap]
        else:
            captions[pid].append(cap)
            
    return captions

captions = readcaptions("../flickr8k/Flickr_Data/Flickr_Data/Flickr_TextData/Flickr8k.token.txt")
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub("[^a-z]+" , " " , text)
    text = text.split()
    
    #text = [ word for word in text if len(word) > 1]
    text = " ".join(text)

    return text
    
    return text
clean_text("My phone number is 94669#####")
```

### ALL Caption cleaning

```python
for key , cap in captions.items():
    for i in range(len(cap)):
        captions[key][i] = clean_text(cap[i])
```

## Creating Vocab

```python
vocab = set()
for key in captions.keys():
    for line in captions[key]:
        vocab.update(line.split())
        
len(vocab)
```

```python
#all words
total_words = []

for key in captions.keys():
    for line in captions[key]:
        total_words.extend(line.split())
        
len(total_words)
```

```python
#Word frequency

word_freq ={}
for word in total_words:
    if(word_freq.get(word) == None):
        word_freq[word] = 1
    else:
        word_freq[word] += 1
        
word_freq
```

```python
new_vocab = set()
for vword in vocab:
    if(word_freq[vword] > 10):
        new_vocab.add(vword)
    
len(new_vocab)
```

```python
def readfiles(path):
    f = open(path)
    text = f.read()
    f.close()
    return text
```

## Seprating Train and Test data

```python
train_data = readfiles("../flickr8k/Flickr_Data/Flickr_Data/Flickr_TextData/Flickr_8k.trainImages.txt")
train = [row for row in train_data.split("\n")[:-1]]
print(len(train))

test_data = readfiles("../flickr8k/Flickr_Data/Flickr_Data/Flickr_TextData/Flickr_8k.testImages.txt")
test = [row for row in test_data.split("\n")[:-1]]
print(len(test))
```

## Tweak -  Adding start and end token to training data 

```python
train_description = {}

for img_id in train:
    train_description[img_id] = []
    for cap in captions[img_id]:
        new_cap = "startseq " + cap + " endseq"
        train_description[img_id].append(new_cap)

train_description[temp_img]
```

```python
from keras.applications.resnet50 import ResNet50
```

```python
model = ResNet50(weights="imagenet" , input_shape= (224,224 ,3))
```

```python
model.summary()
```

```python
from keras.models import Model 
```

```python
model_new = Model(model.input , model.layers[-2].output)
```

```python
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input

path = "../flickr8k/Flickr_Data/Flickr_Data/Images/"

def preprocess_img(img):
    path = "../flickr8k/Flickr_Data/Flickr_Data/Images/"
    img = image.load_img(path + img , target_size=(224 , 224))
    img = image.img_to_array(img)
    img = np.expand_dims(img , axis = 0)
    #Normalisation
    img = preprocess_input(img)
    return img
```

```python
img = preprocess_img(temp_img)
plt.imshow(img[0])
plt.axis("off")
plt.show()
```
